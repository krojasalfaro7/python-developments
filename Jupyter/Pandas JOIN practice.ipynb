```python
data = ps.sqldf("""
        SELECT 
            so.*, arr.*,
            CASE WHEN arr.ARRENDATARIO IS NULL THEN 0 ELSE 1 END AS is_arrendatario
        FROM
            excel_data_socios so
                LEFT JOIN
            excel_data_arrendatario arr ON arr.ARRENDATARIO = so.socio
""")
```
