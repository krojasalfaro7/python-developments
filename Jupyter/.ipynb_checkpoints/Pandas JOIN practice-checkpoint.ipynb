```python
import pandas as pd
import pandasql as ps
```

```python
excel_data_socios = pd.read_excel('data_example.xlsx', sheet_name='Lista_Socios')
excel_data_arrendatario = pd.read_excel('data_example.xlsx', sheet_name='Lista_Arrendatario')
```

```python
excel_data_socios
```

```python
excel_data_arrendatario
```

```python
data = ps.sqldf("""
        SELECT 
            so.*, arr.*,
            CASE WHEN arr.ARRENDATARIO IS NULL THEN 0 ELSE 1 END AS is_arrendatario
        FROM
            excel_data_socios so
                LEFT JOIN
            excel_data_arrendatario arr ON arr.ARRENDATARIO = so.socio
""")
```

```python
data
```
